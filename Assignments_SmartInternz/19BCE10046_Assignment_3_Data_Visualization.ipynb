```python
import random  
import numpy as np  
import pandas as pd  
import seaborn as sns  
import matplotlib.pyplot as plt 
```

# 1. Line Graph

```python
x = np.array([ 0, 0.53, 1.05, 1.58, 2.11, 2.63, 3.16, 3.68, 4.21,
4.74, 5.26, 5.79, 6.32, 6.84])
y = np.tan(x)
```

```python
plt.plot(x,y,'r')
plt.title('tan graph')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

# 2. Scatter Plot





```python
plt.scatter(x,y,marker='s')
plt.title('tan graph')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

# 3. Bar Graph

```python
plt.bar([10,20,30,40,50],[30,50,60,20,40],width=5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

# 4. Histograms

```python
# Let 'x' be the data with 1000 random points.
x = np.random.randn(1000)
```

```python
plt.hist(x,color='green',bins=500)  
plt.xlabel('X-axis')  
plt.ylabel('Y-axis')  
plt.title('Distribution of the variable x')  
plt.show()  
```

# 5. Count Plot

```python
tips=sns.load_dataset('tips') #predefined
tips
```

```python
sns.countplot(x='sex',data=tips)
```

# 6. HeatMap

```python
sns.heatmap(tips.corr(),annot=True)
```

# 7. Box Plot

```python
sns.boxplot(x='size',y='total_bill',data=tips)
```

# 8. Pair Plot

```python
sns.pairplot(tips)
```

# 9. Joint Plot

```python
sns.jointplot(x='size',y='tip',data=tips,kind='hex',color='r')
```

# 10. Kernel density plot

```python
x = np.random.rand(200)   
sns.kdeplot(x, shade=True,color='green')  
```
